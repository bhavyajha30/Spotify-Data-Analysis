```python
##Top 5 most popular artists:
top_five_artists=df.groupby("artist").count().sort_values(by="song_title", ascending=False)["song_title"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

```python
##Top 5 loudest tracks:
top_five_loudest_tracks=df[["loudness", "song_title"]].sort_values(by="loudness", ascending=True)[:5]
top_five_loudest_tracks
```

```python
plt.figure(figsize=(12,7))
x=[-33.097, -31.367, -31.082, -30.447, -29.460]
y=["Lyric Pieces, Book I Op. 12: I. Arietta",	"8 FantasiestÃ¼cke, Op.12 : 1. Des Abends", "Piano Quartet in E flat, Op.47: 3. Andante can...", "I Was So Young, and You Were So Beautiful", "Eleanor"]
colors=["red", "green", "blue", "orange", "purple"]
sns.barplot(x="loudness", y="song_title", hue="song_title", data=top_five_loudest_tracks, dodge=False, palette=colors)
plt.title("Top 5 loudest tracks")
plt.show()
```

```python
#Artist with the most danceability song:
top_artist_danceability=df[["danceability", "song_title", "artist"]].sort_values(by="danceability", ascending=False)[:1]
top_artist_danceability
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability", y="artist", data=top_artist_danceability)
plt.title("Top Danceability Song")
plt.show()
```

```python
#Top 10 instrumentalness tracks:
top_ten_instrumental_tracks=df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness", ascending=False)[:10]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness", data=top_ten_instrumental_tracks, autopct='%1.2f%%', labels=top_ten_instrumental_tracks.song_title)
plt.title("Top 10 instrumentalness tracks")
plt.show()
```

```python
#Multiple feature plots:
interest_feature_cols=["tempo", "loudness", "acousticness", "danceability", "duration_ms", "energy", "instrumentalness", "liveness", "speechiness", "valence"]
for feature_col in interest_feature_cols:
    pos_data=df[df["target"]==1][feature_col]
    neg_data=df[df["target"]==0][feature_col]
    plt.figure(figsize=(12,7))
    sns.histplot(pos_data, bins=30, label="Positive", color="green", alpha=0.6)
    sns.histplot(neg_data, bins=30, label="Negative", color="red", alpha=0.6)
    #Average Lines
    plt.axvline(pos_data.mean(), color="green", linestyle="--", linewidth=2, label="Positive mean")
    plt.axvline(neg_data.mean(), color="red", linestyle="--", linewidth=2, label="Neagtive mean")
    #Labels
    plt.legend(loc="upper right")
    plt.title(f"Positive and Negative Histogram Plot For {feature_col}")
    plt.show()
    
```
